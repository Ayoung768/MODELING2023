# Exercise - Find the average age of eGRID plants

According to `eGRID2021`, what is the average year plants came online?

This would be easy if `eGRID2021` had information on the year plants came online; but that information is not available.

We need to make an assumption: assume the year a plant came **online is equal to the year the oldest generator came online**

Part a) Read the `GEN21` and `PLNT21` sheets from `eGRID2021` and only keep the relative columns.

```python
## your code
df_gen21 = pd.read_excel(io="eGRID2021_data.xlsx",sheet_name="GEN21",skiprows=1,keep_default_na=False)
df_plant21 = pd.read_excel(io="eGRID2021_data.xlsx",sheet_name="PLNT21",skiprows=1,keep_default_na=False)


df_gen21_sub = df_gen21.loc[:,["ORISPL", "GENYRONL","GENSTAT"]]
df_plant21_sub = df_plant21.loc[:,["ORISPL", "PLPRMFL", "NAMEPCAP", "CAPFAC"]]
```

Part b) Write a code that merges the two DataFrames based on the facility code.

```python
## your code
merged_inner = pd.merge(df_plant21_sub, df_gen21_sub,on="ORISPL")
merged_inner["GENYRONL"] = pd.to_numeric(merged_inner['GENYRONL'])
merged_inner.head()
```

Part c) Now we need to group the rows based on the `ORISPL` column.

```python
## your code
merged_inner_group = merged_inner.groupby(by="ORISPL")
merged_inner_group.head()
```

Part d) Get the minimum year from the column `GENYRONL` in each group.

```python
## your code
date_plant_online = merged_inner_group[["GENYRONL"]].min()
date_plant_online.head(20)
```

Present a table that shows the total number of generators that came online in each state of the U.S. for years 2011, 2012, ..., and 2021. The table should have one row for each of the 52 states (i.e., 50 states + DC and Puerto Rico) in the U.S.

```python
sub_egrid=df_gen21.loc[:,['PSTATABB','PNAME','FUELG1','CFACT','GENYRONL']]
sub_egrid["GENYRONL"] = pd.to_numeric(sub_egrid['GENYRONL'])
sub_egrid=sub_egrid[ (sub_egrid.GENYRONL >= 2011) & (sub_egrid.GENYRONL <= 2021) ]
sub_egrid.shape
```

If you want the years to drop the decimal part and be shown as integer, we can use the `downcast` option of `pd.to_numeric()`

```python
sub_egrid['GENYRONL']=pd.to_numeric(sub_egrid['GENYRONL'],downcast='integer')
sub_egrid.head()
```

```python
pivot1=sub_egrid.pivot_table(values="PNAME",index="PSTATABB",columns="GENYRONL",aggfunc="count",margins=True)
pivot1.head(10)
```

```python
pivot1=sub_egrid.pivot_table(values="PNAME",index="PSTATABB",columns="GENYRONL",aggfunc="count",fill_value="",margins=True)
pivot1.head(52)
```

```python
sub_egrid_plant=df_plant21.loc[:,['PSTATABB','PNAME','PLFUELCT']]

pivot2=sub_egrid_plant.pivot_table(values="PNAME",index="PSTATABB",columns="PLFUELCT",aggfunc="count",fill_value="",margins=True)
pivot2.head()
```

We want to merge the two pivot tables. The default command `pd.merge(t1,t2)` does not work; we need to specify the column to use for merging.

```python
finaldf = pd.merge(pivot1, pivot2,on="PSTATABB")
finaldf = pivot1.merge(pivot2, on = "PSTATABB")
finaldf.head()
```

A different way to combine the two pivot tables is by concatenating them. The `axis` parameter specifies the direction of concatenation.

```python
df_concat = pd.concat( [pivot1, pivot2], axis = 0 )
df_concat.head(100)
```

```python
df_concat = pd.concat( [pivot1, pivot2], axis = 1 )
df_concat.head()
```
